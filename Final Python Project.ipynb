```python
import numpy as np
dt['death_rate'] = np.where(
    dt['total_cases'] > 0,
    dt['total_deaths'] / dt['total_cases'],
    0
)
dt['death_rate_percent'] = dt['death_rate'] * 100
print(dt[['total_cases', 'total_deaths', 'death_rate_percent']])
```

```python
countries = ['South Africa', 'Kenya', 'Lesotho']
dt_filtered = dt[dt['location'].isin(['South Africa', 'Kenya', 'Lesotho'])]

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

for country in countries:
    subset = dt_filtered[dt_filtered['location'] == country]
    plt.plot(subset['date'], subset['new_cases'], label=country)

plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.title('Daily COVID-19 New Cases by Country')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

latest_data = dt_filtered.sort_values('date').groupby('location').last().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(latest_data['location'], latest_data['total_cases'], color='mediumseagreen')

plt.xlabel('location')
plt.ylabel('Total Cases')
plt.title('Total COVID-19 Cases: Kenya, South Africa, and Lesotho')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
dt_filtered = dt_filtered.sort_values(by=['location', 'date'])
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = dt_filtered[dt_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=f'{country} - Cases', linestyle='-', marker='o')
    plt.plot(country_data['date'], country_data['total_deaths'], label=f'{country} - Deaths', linestyle='--', marker='x')
plt.xlabel('Date')
plt.ylabel('Total Cases / Total Deaths')
plt.title('COVID-19 Cases & Deaths Over Time: Kenya, South Africa, Lesotho')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dt_filtered = dt_filtered.sort_values(by=['location', 'date'])

plt.figure(figsize=(12, 6))

for country in countries:
    country_data = dt_filtered[dt_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=f'{country} - Vaccinations', marker='o')

# Step 4: Add labels and title
plt.xlabel('Date')
plt.ylabel('Cumulative Vaccinations')
plt.title('Cumulative COVID-19 Vaccinations Over Time: Kenya, South Africa, Lesotho')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
population_data = {
    'Kenya': 	57532493,      
    'South Africa': 64747319,  
    'Lesotho': 2359878   }   
dt['total_population'] = dt['location'].map(population_data)
countries = ['Kenya', 'South Africa', 'Lesotho']
dt_filtered = dt[dt['location'].isin(countries)].sort_values(by=['location', 'date'])
dt_filtered['vaccinated_percentage'] = (dt_filtered['total_vaccinations'] / dt_filtered['total_population']) * 100
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = dt_filtered[dt_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['vaccinated_percentage'], label=f'{country} - Vaccinated (%)', marker='o')
plt.xlabel('Date')
plt.ylabel('Vaccinated Population (%)')
plt.title('Percentage of Vaccinated Population Over Time: Kenya, South Africa, Lesotho')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
South Africa has the highest vaccinated population compared to Kenya and Lesotho.
South Africa has the highest COVID cases, with Kenya coming in second.
```
